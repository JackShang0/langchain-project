举例1：从物理磁盘的yaml文件中加载Prompt模版

```python
from langchain_core.prompts import load_prompt
from dotenv import load_dotenv

load_dotenv()
prompt = load_prompt("prompt/prompt.yaml", encoding="utf-8")
# print(prompt)
print(prompt.format(name="年轻人", what="滑稽"))
```
