## 1.3、 关于模型调用的方法
为了尽可能简化自定义链的创建，我们实现了一个Runnable协议。许多LangChain组件实现了Runnable，包括聊天模型、提示词模板、输出解析器、检索器、代理等
## Runnable定义的公共的调用方法如下：
- invoke：处理单条输入，等到LLM完全推理完成后再返回调用结果
- stream：流式响应，逐字输出LLM的响应结果
- batch：处理批量输入

这些方法也有相应的异步方法，应该与 asyncio 的 await 语法一起使用以实现并发：
- astream：异步流式响应
- ainvoke：异步处理单条输入
- abatch：异步处理批量输入
- astream_log：异步流式返回中间步骤，以及最终响应
- astream_events：

## 1.3.1 流式输出与非流式输出
在LangChain中，语言模型的输出分为两种模式：流式输出与非流式输出
下面是两个场景：
- 非流式输出：这是LangChain与LLM交互时的默认行为，是最简单、最稳定的语言模型调用方式。当用户发出请求后，系统在后台等待模型生成完整响应，然后一次性将全部结果返回。
- 流式输出：一种更具交互感的模型输出方式，用户不再需要等待完整答案，而是能看到模型逐个token地实时返回内容。

流式输出演示

```python
from langchain_core.messages import HumanMessage, SystemMessage

system_message = SystemMessage(
    content="我是一个人工智能助手，我的名字叫小智",
)

human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [system_message, human_message]
# print(messages)


import dotenv
from langchain_openai import ChatOpenAI
import os

dotenv.load_dotenv()  # 加载当前目录下的 .env 文件
os.environ.setdefault("OPENAI_API_KEY", os.getenv("OPENAI_API_KEY"))
os.environ.setdefault("OPENAI_BASE_URL", os.getenv("OPENAI_BASE_URL"))

# 获取对话模型
model = ChatOpenAI(
    model="deepseek-chat",
    streaming=True  #流式输出关键配置
)
print("开始流式输出")
for chunk in model.stream(messages):
    # 逐个打印内容
    print(chunk.content, end="", flush=True)        # 逐个打印流式输出结果

print("\n流式输出结束")
```

## 1.3.2、批量调用

批量调用举例

```python
from langchain_core.messages import HumanMessage, SystemMessage
import dotenv
import os
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()  # 加载当前目录下的 .env 文件
os.environ.setdefault("OPENAI_API_KEY", os.getenv("OPENAI_API_KEY"))
os.environ.setdefault("OPENAI_BASE_URL", os.getenv("OPENAI_BASE_URL"))

# 获取对话模型
model = ChatOpenAI(
    model="deepseek-chat",
    streaming=True  #关键配置
)

message1 =  [SystemMessage(content="你是一个位乐于助人的智能小助手"),HumanMessage(content="请帮我介绍一下什么是机器学习")]
message2 =  [SystemMessage(content="你是一个位乐于助人的智能小助手"),HumanMessage(content="请帮我介绍一下什么是AIGC")]
message3 =  [SystemMessage(content="你是一个位乐于助人的智能小助手"),HumanMessage(content="请帮我介绍一下什么是大模型技术")]

messages = [message1, message2, message3]

response = model.batch(messages)
print(response)

print("开始流式输出")
for chunk in model.batch(messages):
    # 逐个打印内容
    print(chunk.content, end="", flush=True)

print("\n流式输出结束")
```

# 1.3.3、关于同步与异步方法调用

```python
import time
import asyncio

# 异步执行

async def async_call():
    await asyncio.sleep(5)  #模拟异步任务
    print("异步调用成功")


async def perform_other_tasks():
    await asyncio.sleep(5)
    print("其他任务完成")


async def run_async_tasks():
    start_time = time.time()
    await asyncio.gather(
        async_call(),
        perform_other_tasks()
    )
    end_time = time.time()
    return f"总耗时：{end_time - start_time}"


if __name__ == '__main__':
    result = await run_async_tasks()
    print(result)
```
